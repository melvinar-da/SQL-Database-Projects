```python
import pandas as pd
from google.colab import files
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Upload CSV from local machine (if needed)
uploaded = files.upload()

# Load dataset
df = pd.read_csv("customer_sales.csv")

# Check data
df.head()

# Remove missing values
df = df.dropna()

# Keep relevant columns
df = df[['CustomerID', 'TotalSpent']]

# Normalize data
df['TotalSpent'] = (df['TotalSpent'] - df['TotalSpent'].min()) / (df['TotalSpent'].max() - df['TotalSpent'].min())

df.head()

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Run K-Means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['TotalSpent']])

# Display results
df.head()


# Plot customer spending clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['CustomerID'], df['TotalSpent'], c=df['Cluster'], cmap='viridis')
plt.xlabel("Customer ID")
plt.ylabel("Total Spent")
plt.title("Customer Segments Based on Spending")
plt.colorbar(label="Cluster")
plt.show()
```
